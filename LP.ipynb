```python
import numpy as np
import matplotlib.pyplot as plt

# Define the constraint lines
x = np.linspace(0, 5, 400)  # Range for plotting

# Calories constraint: 200x1 + 150x2 + 300x3 >= 500 (x3 = 0 for 2D)
calories_y = (500 - 200 * x) / 150

# Protein constraint: 10x1 + 5x2 + 20x3 >= 25 (x3 = 0 for 2D)
protein_y = (25 - 10 * x) / 5

# Feasible region
plt.figure(figsize=(8, 6))
plt.plot(x, calories_y, label="Calories Constraint", linestyle='--')
plt.plot(x, protein_y, label="Protein Constraint", linestyle='--')

# Highlight feasible region
plt.fill_between(x, np.maximum(calories_y, protein_y), 10, color='gray', alpha=0.2, label="Feasible Region")

# Mark the optimal solution
optimal_x1 = 0  # From the result
optimal_x2 = 1.67
plt.scatter(optimal_x1, optimal_x2, color='red', zorder=5, label="Optimal Solution (0, 1.67)")

# Labels and legend
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.xlabel("Quantity of Food A (x1)")
plt.ylabel("Quantity of Food B (x2)")
plt.title("Diet Problem - Feasible Region and Constraints")
plt.axhline(0, color="black",linewidth=0.5)
plt.axvline(0, color="black",linewidth=0.5)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

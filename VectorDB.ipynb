```python
import sqlite3
import numpy as np
from scipy.spatial.distance import cosine, euclidean, cityblock, chebyshev, mahalanobis

# Connect to SQLite and create a table for vector storage (simulating a vector DB)
conn = sqlite3.connect(':memory:')  # In-memory database for simplicity
cur = conn.cursor()

# Create table for storing vectors
cur.execute('''
CREATE TABLE vectors (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    vector BLOB
)
''')

# Convert vector to binary blob for storage


def vector_to_blob(vector):
    return np.array(vector).tobytes()

# Convert binary blob back to vector


def blob_to_vector(blob):
    return np.frombuffer(blob, dtype=np.float64)


# Simulate inserting 100 vectors into the "vector DB"
np.random.seed(42)
vectors = np.random.randn(100, 3)

# Insert vectors into the database
for vector in vectors:
    cur.execute("INSERT INTO vectors (vector) VALUES (?)",
                (vector_to_blob(vector),))

conn.commit()

# CRUD Operations


def insert_vector(vector):
    cur.execute("INSERT INTO vectors (vector) VALUES (?)",
                (vector_to_blob(vector),))
    conn.commit()


def update_vector(vector_id, new_vector):
    cur.execute("UPDATE vectors SET vector = ? WHERE id = ?",
                (vector_to_blob(new_vector), vector_id))
    conn.commit()


def delete_vector(vector_id):
    cur.execute("DELETE FROM vectors WHERE id = ?", (vector_id,))
    conn.commit()


def get_vector(vector_id):
    cur.execute("SELECT vector FROM vectors WHERE id = ?", (vector_id,))
    row = cur.fetchone()
    if row:
        return blob_to_vector(row[0])
    return None

# Example of querying and distance calculation


def query_similar_vectors(target_vector, distance_metric='euclidean', top_n=5):
    cur.execute("SELECT id, vector FROM vectors")
    all_vectors = cur.fetchall()

    distances = []
    for vector_id, vector_blob in all_vectors:
        vector = blob_to_vector(vector_blob)

        # Calculate distance based on the chosen metric
        if distance_metric == 'euclidean':
            dist = euclidean(target_vector, vector)
        elif distance_metric == 'cosine':
            dist = cosine(target_vector, vector)
        elif distance_metric == 'manhattan':
            dist = cityblock(target_vector, vector)
        elif distance_metric == 'chebyshev':
            dist = chebyshev(target_vector, vector)
        else:
            raise ValueError(f"Unsupported distance metric: {distance_metric}")

        distances.append((vector_id, dist))

    # Sort by distance and return the top N closest vectors
    distances.sort(key=lambda x: x[1])
    return distances[:top_n]


# Insert a new vector
new_vector = np.random.randn(3)
insert_vector(new_vector)

# Update a vector
update_vector(1, np.array([0.5, 0.5, 0.5]))

# Delete a vector
delete_vector(2)

# Retrieve a vector
vector = get_vector(1)
print("Retrieved vector:", vector)

# Query the top 5 similar vectors based on Euclidean distance
target_vector = np.array([1, 1, 1])
similar_vectors = query_similar_vectors(
    target_vector, distance_metric='euclidean', top_n=5)
print("\nTop 5 similar vectors based on Euclidean distance:")
for vector_id, dist in similar_vectors:
    print(f"Vector ID: {vector_id}, Distance: {dist}")

# Close the connection
conn.close()
```

```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph to visualize the database schema with cardinality and attributes
G = nx.DiGraph()

# Define the tables
tables = [
    "users", "patients", "therapists", "exercises", "sessions", "session_exercises",
    "camera_tracking", "progress_reports", "appointments", "body_markers", 
    "session_body_marker_tracking", "video_files", "logs"
]

# Define the relationships
relationships = [
    ("patients", "users"), # patients -> users (Many-to-One)
    ("therapists", "users"), # therapists -> users (Many-to-One)
    ("sessions", "patients"), # sessions -> patients (Many-to-One)
    ("sessions", "therapists"), # sessions -> therapists (Many-to-One)
    ("session_exercises", "sessions"), # session_exercises -> sessions (Many-to-One)
    ("session_exercises", "exercises"), # session_exercises -> exercises (Many-to-One)
    ("camera_tracking", "sessions"), # camera_tracking -> sessions (Many-to-One)
    ("progress_reports", "patients"), # progress_reports -> patients (Many-to-One)
    ("progress_reports", "users"), # progress_reports -> users (Many-to-One, generated_by)
    ("appointments", "patients"), # appointments -> patients (Many-to-One)
    ("appointments", "therapists"), # appointments -> therapists (Many-to-One)
    ("session_body_marker_tracking", "sessions"), # session_body_marker_tracking -> sessions (Many-to-One)
    ("session_body_marker_tracking", "body_markers"), # session_body_marker_tracking -> body_markers (Many-to-One)
    ("video_files", "sessions"), # video_files -> sessions (Many-to-One)
    ("logs", "users") # logs -> users (Many-to-One)
]

# Add nodes (tables) to the graph
G.add_nodes_from(tables)

# Add relationships (edges) to the graph
G.add_edges_from(relationships)

# Define cardinalities
cardinalities = {
    ("patients", "users"): "1-to-Many",
    ("therapists", "users"): "1-to-Many",
    ("sessions", "patients"): "Many-to-1",
    ("sessions", "therapists"): "Many-to-1",
    ("session_exercises", "sessions"): "Many-to-1",
    ("session_exercises", "exercises"): "Many-to-1",
    ("camera_tracking", "sessions"): "Many-to-1",
    ("progress_reports", "patients"): "Many-to-1",
    ("progress_reports", "users"): "Many-to-1",
    ("appointments", "patients"): "Many-to-1",
    ("appointments", "therapists"): "Many-to-1",
    ("session_body_marker_tracking", "sessions"): "Many-to-1",
    ("session_body_marker_tracking", "body_markers"): "Many-to-1",
    ("video_files", "sessions"): "Many-to-1",
    ("logs", "users"): "Many-to-1"
}


# Define the attributes for each table
attributes = {
    "users": ["user_id", "role", "first_name", "last_name", "email", "password_hash", "created_at"],
    "patients": ["patient_id", "user_id", "date_of_birth", "medical_history", "assigned_therapist_id"],
    "therapists": ["therapist_id", "user_id", "specialization"],
    "exercises": ["exercise_id", "name", "description", "target_limb", "difficulty_level"],
    "sessions": ["session_id", "patient_id", "therapist_id", "session_date", "notes"],
    "session_exercises": ["session_exercise_id", "session_id", "exercise_id", "repetitions", "sets"],
    "camera_tracking": ["tracking_id", "session_id", "limb", "movement_accuracy", "movement_range", "timestamp"],
    "progress_reports": ["report_id", "patient_id", "report_date", "summary", "generated_by"],
    "appointments": ["appointment_id", "patient_id", "therapist_id", "appointment_date", "status", "created_at"],
    "body_markers": ["marker_id", "name", "description"],
    "session_body_marker_tracking": ["tracking_id", "session_id", "marker_id", "position_x", "position_y", "position_z", "timestamp"],
    "video_files": ["video_id", "session_id", "file_path", "uploaded_at"],
    "logs": ["log_id", "user_id", "action", "action_time"]
}

# Create labels with attributes for each table
node_labels = {table: f"{table}\n" +
               "\n".join(attrs) for table, attrs in attributes.items()}

# Adjust the spring layout to increase the distance between nodes
pos = nx.spring_layout(G, k=1.5)

# Draw the graph with table names and attributes
plt.figure(figsize=(14, 10))
nx.draw(G, pos, with_labels=False, node_size=3000,
        node_color="lightblue", font_size=10, font_weight="bold", arrows=True)

# Draw edges with cardinalities
edge_labels = {(u, v): cardinalities.get((u, v), '') for u, v in G.edges()}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')

# Add the labels (table names + attributes)
nx.draw_networkx_labels(G, pos, labels=node_labels, font_size=8)

plt.title("Database Schema Relationship Diagram (with Cardinality and Attributes)", fontsize=15)
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Initialize the directed graph
G = nx.DiGraph()

# Define the tables and attributes
tables = {
    "Patients": ["patient_id (PK)", "name", "email", "date_of_birth", "phone_number", "condition_id (FK)", "created_at"],
    "Physiotherapists": ["physiotherapist_id (PK)", "name", "email", "phone_number", "specialization", "created_at"],
    "Conditions": ["condition_id (PK)", "condition_name", "description"],
    "Exercises": ["exercise_id (PK)", "exercise_name", "description", "condition_id (FK)", "physiotherapist_id (FK)"],
    "Videos": ["video_id (PK)", "video_url", "exercise_id (FK)"],
    "Appointments": ["appointment_id (PK)", "patient_id (FK)", "physiotherapist_id (FK)", "appointment_date", "status"],
    "ExerciseSessions": ["session_id (PK)", "patient_id (FK)", "exercise_id (FK)", "session_date", "progress_id (FK)"],
    "Progress": ["progress_id (PK)", "session_id (FK)", "feedback", "progress_percentage"],
    "LimbAngles": ["angle_id (PK)", "session_id (FK)", "limb_name", "angle_value", "is_correct_angle"]
}

# Add nodes to the graph (tables)
for table, attributes in tables.items():
    G.add_node(table, label="\n".join(attributes))

# Define the relationships and cardinalities
relationships = [
    ("Patients", "Appointments", "1:N"),
    ("Physiotherapists", "Appointments", "1:N"),
    ("Patients", "ExerciseSessions", "1:N"),
    ("Exercises", "ExerciseSessions", "1:N"),
    ("Conditions", "Patients", "1:N"),
    ("Conditions", "Exercises", "1:N"),
    ("Physiotherapists", "Exercises", "1:N"),
    ("Exercises", "Videos", "1:N"),
    ("ExerciseSessions", "Progress", "1:1"),
    ("ExerciseSessions", "LimbAngles", "1:N"),
    ("Progress", "ExerciseSessions", "1:1")
]

# Add edges to the graph (relationships)
for (source, target, cardinality) in relationships:
    G.add_edge(source, target, label=cardinality)

# Set a larger figure size
plt.figure(figsize=(14, 10))  # Width and height in inches

# Manually specify positions to centralize "Patients" and "Physiotherapists"
pos = {
    "Patients": (0, 0),
    "Physiotherapists": (1, 0),
    "Conditions": (-1.5, 1.5),
    "Exercises": (1.5, 1.5),
    "Videos": (2.5, 2.5),
    "Appointments": (-1.5, -1.5),
    "ExerciseSessions": (1.5, -1.5),
    "Progress": (2.5, -2.5),
    "LimbAngles": (3.5, -3.5)
}

# Draw nodes with table names and attributes
nx.draw_networkx_nodes(G, pos, node_size=7000, node_color="skyblue", node_shape="s")
nx.draw_networkx_labels(G, pos, labels={node: f"{node}\n" + "\n".join(tables[node]) for node in G.nodes()}, font_size=9)

# Draw edges with cardinalities
nx.draw_networkx_edges(G, pos, arrowstyle="->", arrowsize=20, edge_color="black", width=2)
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): d['label'] for u, v, d in G.edges(data=True)}, font_color='red')

# Show the plot
plt.title("Database Schema and Relationships")
plt.axis("off")
plt.show()
```

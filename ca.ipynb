```python
import pandas as pd
import prince
import matplotlib.pyplot as plt

# Load the CSV file
# Replace "your_file.csv" with the path to your CSV file
df = pd.read_csv("spese.csv")

# Create a contingency table for CAUSALE and TIPO
contingency_table = pd.crosstab(df['CAUSALE'], df['TIPO'])

# Perform Correspondence Analysis with a specified engine
ca = prince.CA(n_components=2, n_iter=10, copy=True,
               check_input=True, engine='scipy')
ca = ca.fit(contingency_table)

# Get row and column coordinates
row_coords = ca.row_coordinates(contingency_table)
col_coords = ca.column_coordinates(contingency_table)

# Plot the results
plt.figure(figsize=(10, 8))

# Plot row coordinates (CAUSALE)
plt.scatter(row_coords[0], row_coords[1], color='red', label='CAUSALE')
for i, label in enumerate(row_coords.index):
    plt.text(row_coords[0][i], row_coords[1][i], label, color='red')

# Plot column coordinates (TIPO)
plt.scatter(col_coords[0], col_coords[1], color='blue', label='TIPO')
for i, label in enumerate(col_coords.index):
    plt.text(col_coords[0][i], col_coords[1][i], label, color='blue')

# Add labels and legend
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.title('Correspondence Analysis of CAUSALE and TIPO')
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from prince import CA  # Correspondence Analysis library

# Sample survey data: rows represent customer segments, columns represent preferences
data = {
    "Segment A": [30, 10, 20, 5],
    "Segment B": [20, 25, 15, 10],
    "Segment C": [5, 15, 25, 20],
    "Segment D": [10, 5, 10, 30],
}
df = pd.DataFrame(data, index=["Product X", "Product Y", "Product Z", "Product W"])

# Perform Correspondence Analysis
ca = CA(n_components=2)  # Reduce to 2 dimensions for easy visualization
ca.fit(df)

# Get transformed data
row_coords = ca.row_coordinates(df)
col_coords = ca.column_coordinates(df)

print("Row Coordinates:\n", row_coords)
print("Column Coordinates:\n", col_coords)

# Visualize the results
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))
# Plot row points
plt.scatter(row_coords[0], row_coords[1], label="Products", color="blue")
for i, label in enumerate(df.index):
    plt.text(row_coords[0][i], row_coords[1][i], label, color="blue")

# Plot column points
plt.scatter(col_coords[0], col_coords[1], label="Segments", color="red")
for i, label in enumerate(df.columns):
    plt.text(col_coords[0][i], col_coords[1][i], label, color="red")

plt.axhline(0, color="gray", linestyle="--", linewidth=0.5)
plt.axvline(0, color="gray", linestyle="--", linewidth=0.5)
plt.legend()
plt.title("Correspondence Analysis: Hidden Patterns Revealed!")
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from prince import CA  # Correspondence Analysis library
import numpy as np
import matplotlib.pyplot as plt

# Generate a larger dataset: 10 products and 6 customer segments
np.random.seed(42)
rows = [f"Product {chr(65+i)}" for i in range(10)]  # Products A-J
columns = [f"Segment {chr(65+i)}" for i in range(6)]  # Segments A-F

data = np.random.randint(10, 100, size=(10, 6))  # Random preferences data
df = pd.DataFrame(data, index=rows, columns=columns)

# Perform Correspondence Analysis
ca = CA(n_components=2)  # Reduce to 2 dimensions for visualization
ca.fit(df)

# Get transformed data
row_coords = ca.row_coordinates(df)
col_coords = ca.column_coordinates(df)

# Visualization
plt.figure(figsize=(12, 8))

# Plot row points (Products)
plt.scatter(row_coords[0], row_coords[1], color="blue", label="Products")
for i, label in enumerate(df.index):
    plt.text(row_coords[0][i], row_coords[1][i], label, color="blue", fontsize=10)

# Plot column points (Segments)
plt.scatter(col_coords[0], col_coords[1], color="red", label="Segments")
for i, label in enumerate(df.columns):
    plt.text(col_coords[0][i], col_coords[1][i], label, color="red", fontsize=10)

# Add auxiliary lines
plt.axhline(0, color="gray", linestyle="--", linewidth=0.5)
plt.axvline(0, color="gray", linestyle="--", linewidth=0.5)

# Final touches
plt.title("Correspondence Analysis: Hidden Patterns in Preferences", fontsize=14)
plt.xlabel("Dimension 1", fontsize=12)
plt.ylabel("Dimension 2", fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
import pandas as pd
from prince import MCA  # Multiple Correspondence Analysis library
import numpy as np
import matplotlib.pyplot as plt

# Generate a larger dataset: 10 products and 6 customer segments with categorical responses
np.random.seed(42)
rows = [f"Product {chr(65+i)}" for i in range(10)]  # Products A-J
columns = [f"Segment {chr(65+i)}" for i in range(6)]  # Segments A-F

# Simulating data where each entry is a categorical choice (e.g., Low, Medium, High)
categories = ['Low', 'Medium', 'High']
data = np.random.choice(categories, size=(10, 6), p=[0.3, 0.4, 0.3])  # Random categorical data
df = pd.DataFrame(data, index=rows, columns=columns)

# Perform Multiple Correspondence Analysis
mca = MCA(n_components=2)  # Reduce to 2 dimensions for visualization
mca.fit(df)

# Get transformed data for rows and columns
row_coords = mca.row_coordinates(df)
col_coords = mca.column_coordinates(df)

# Visualization
plt.figure(figsize=(12, 8))

# Plot row points (Products)
plt.scatter(row_coords[0], row_coords[1], color="blue", label="Products")
for i, label in enumerate(df.index):
    plt.text(row_coords[0][i], row_coords[1][i], label, color="blue", fontsize=10)

# Plot column points (Segments)
plt.scatter(col_coords[0], col_coords[1], color="red", label="Segments")
for i, label in enumerate(df.columns):
    plt.text(col_coords[0][i], col_coords[1][i], label, color="red", fontsize=10)

# Add auxiliary lines
plt.axhline(0, color="gray", linestyle="--", linewidth=0.5)
plt.axvline(0, color="gray", linestyle="--", linewidth=0.5)

# Final touches
plt.title("Multiple Correspondence Analysis: Hidden Patterns in Preferences", fontsize=14)
plt.xlabel("Dimension 1", fontsize=12)
plt.ylabel("Dimension 2", fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Generate a DataFrame with 1000 elements between 140 and 200
data = np.random.uniform(140, 200, 10000)
print(data)
type(data)
```

```python

df = pd.DataFrame({'Values': data})

# Step 2: Create 100 samples of 20 elements each
samples = [df.sample(n=20, replace=False)['Values'].values for _ in range(100)]

# Step 3: Compute the average for each sample
averages = [np.mean(sample) for sample in samples]

# Step 4: Plot the results
plt.figure(figsize=(10, 5))
plt.plot(range(1, 101), averages, marker='o', linestyle='-', color='b', label='Sample Averages')
plt.axhline(y=np.mean(averages), color='r', linestyle='--', label='Overall Mean')
plt.xlabel('Sample Number')
plt.ylabel('Average Value')
plt.title('Sample Averages from 1000 Data Points')
plt.legend()
plt.grid()
plt.show()
```

```python

# Step 5: Plot the sample frequency distribution
plt.figure(figsize=(10, 5))
plt.hist(averages, bins=10, color='c', edgecolor='black', alpha=0.7)
plt.xlabel('Average Value')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of Sample Averages')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
